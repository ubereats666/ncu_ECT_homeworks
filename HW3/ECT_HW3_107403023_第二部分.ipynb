```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('customer_churn.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
df.info()
```

```python
df['Tenure'] = df['Tenure'].fillna(round(df['Tenure'].mean(), 1))
df['WarehouseToHome'] = df['WarehouseToHome'].fillna(round(df['WarehouseToHome'].mean(), 1))
df['HourSpendOnApp'] = df['HourSpendOnApp'].fillna(round(df['HourSpendOnApp'].mean(), 1))
df['OrderAmountHikeFromlastYear'] = df['OrderAmountHikeFromlastYear'].fillna(round(df['OrderAmountHikeFromlastYear'].mean(), 1))
df['CouponUsed'] = df['CouponUsed'].fillna(round(df['CouponUsed'].mean(), 1))
df['OrderCount'] = df['OrderCount'].fillna(round(df['OrderCount'].mean(), 1))
df['DaySinceLastOrder'] = df['DaySinceLastOrder'].fillna(round(df['DaySinceLastOrder'].mean(), 1))

df.info()
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df["PreferredLoginDevice"] = obj_df["PreferredLoginDevice"].astype('category')
obj_df["PreferredPaymentMode"] = obj_df["PreferredPaymentMode"].astype('category')
obj_df["Gender"] = obj_df["Gender"].astype('category')
obj_df["PreferedOrderCat"] = obj_df["PreferedOrderCat"].astype('category')
obj_df["MaritalStatus"] = obj_df["MaritalStatus"].astype('category')
obj_df.dtypes
```

```python
obj_df["PreferredLoginDevice_cat"] = obj_df["PreferredLoginDevice"].cat.codes
obj_df["PreferredPaymentMode_cat"] = obj_df["PreferredPaymentMode"].cat.codes
obj_df["Gender_cat"] = obj_df["Gender"].cat.codes
obj_df["PreferedOrderCat_cat"] = obj_df["PreferedOrderCat"].cat.codes
obj_df["MaritalStatus_cat"] = obj_df["MaritalStatus"].cat.codes
obj_df
```

```python
df['PreferredLoginDevice'] = obj_df['PreferredLoginDevice_cat']
df['PreferredPaymentMode'] = obj_df['PreferredPaymentMode_cat']
df['Gender'] = obj_df['Gender_cat']
df['PreferedOrderCat'] = obj_df['PreferedOrderCat_cat']
df['MaritalStatus'] = obj_df['MaritalStatus_cat']
df
```

```python
X = df.drop(['Churn'], axis =1).values
y = df['Churn'].values

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=12)
```

```python
#SVM 模型

from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

svm = make_pipeline(StandardScaler(), SVR(kernel='linear', degree=1, gamma='auto',C=1.0, epsilon=0.2))
```

```python
from sklearn.svm import SVR
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_score = svm.score(X_test, y_test)
svm_score
```

```python
#Logistic Regression 模型

from sklearn import preprocessing, linear_model

logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X_train, y_train)
logistic_pred = logistic_regr.predict(X_test)
logistic_score = logistic_regr.score(X_test, y_test)
logistic_score
```

```python
# Decision tree 模型

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
tree_pred = clf.predict(X_test)
tree_score = clf.score(X_test, y_test)
tree_score
```

```python
svm_pred
```

```python
# meanings of hyperparameter, in SVM model,
# c = penalty for each misclassified data point
# gamma = controls the distance of influence of a single training point
```

```python
# 請以 Accuracy 評估同一模型在不同超參數設定下於訓練資料及測試資料上的表現
# 測試SVM模型中 C = 1, 10, 100的score

svm_10 = make_pipeline(StandardScaler(), SVR(kernel='linear', degree=1, gamma='auto',C=10.0, epsilon=0.2))
svm_10.fit(X_train, y_train)
svm_pred_10 = svm_10.predict(X_test)
svm_score_10 = svm_10.score(X_test, y_test)
svm_score_10
```

```python
svm_100 = make_pipeline(StandardScaler(), SVR(kernel='linear', degree=1, gamma='auto',C=100.0, epsilon=0.2))
svm_100.fit(X_train, y_train)
svm_pred_100 = svm_100.predict(X_test)
svm_score_100 = svm_100.score(X_test, y_test)
svm_score_100
```

```python
compare = {
    "SVM model": ["c = 1", "c = 10", "c = 100"],
    "score": [svm_score, svm_score_10, svm_score_100],
}

compare_df = pd.DataFrame(compare)
compare_df
```

```python
#請以 Accuracy 比較 三個模型的表現 

score_dict = {
    "model" : ["SVM", "logistic regression", "decision tree"],
    "score" : [svm_score, logistic_score, tree_score]
}

score_df = pd.DataFrame(score_dict)
score_df
```
